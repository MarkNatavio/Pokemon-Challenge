# Introduction

This is my submission for Question 4. Here, I have created a multiple regression prediction model to determine if a Pokemon is a Legendary Pokemon or not.

```python
# Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score

# load dataframe
df = pd.read_csv("/Users/marknatavio/Desktop/Challenge/pokemon_data_science.csv")
df
```

# Feature Processing

```python
# Lets load up the dependent and independent values

# remove all those fields we don't need or want to test with
x = df.drop(['isLegendary','Name','Type_1','Type_2','Color','Pr_Male','Egg_Group_1','Egg_Group_2','Catch_Rate','Body_Style' ], axis = 1).values

# save target value
y = df['isLegendary'].values
```

```python
# Check that x works
#print(x)

# Check that y works
#print(y)
```

```python
# Get the testing and training values for x and y. I will be using 21 rows for teting and 700 for training
# The more training the better

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 21, train_size = 700)
```

# Model Fitting

```python
# Fitting the model
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# Testing the prediction
pred = reg.predict(x_test)
pred
```

# Results & Model Evaluation

```python
clf = LogisticRegression(random_state=0).fit(x_train, y_train)

# The coefficients
print("Coefficients: \n", clf.coef_)
# Score
print("\nScore:",clf.score(x_test, y_test))
# Cross_validating
print("Score gotten from cross-validation:",reg.score(x_test,y_test))
```
